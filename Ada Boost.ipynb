```python
df1=df.drop('class',axis=1)
x=pd.get_dummies(df1,drop_first=True)
y=df['class']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
# model with default parameters
```

```python
from sklearn.ensemble import AdaBoostClassifier
model=AdaBoostClassifier()
model.fit(x_train,y_train)

# predictions
train_pred=model.predict(x_train)
test_pred=model.predict(x_test)

#evaluation
from sklearn.metrics import accuracy_score
print('train accuracy:',accuracy_score(y_train,train_pred))
print('test accuracy:',accuracy_score(y_test,test_pred))

#cross val score
from sklearn.model_selection import cross_val_score
score=cross_val_score(model,x,y,cv=5)
print('cross validation score:',score.mean())
```

```python
#hyperparameter tuning

from sklearn.model_selection import GridSearchCV
estimator=AdaBoostClassifier()
param_grid={'n_estimators':list(range(1,51))}

grid=GridSearchCV(estimator,param_grid,scoring='accuracy',cv=5)
grid.fit(x_train,y_train)
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid.best_estimator_.feature_importances_
```

```python
feats=pd.DataFrame(index=x.columns,data=grid.best_estimator_.feature_importances_,columns=['feature importance'])
```

```python
important_features=feats[feats['feature importance']>0]
important_features
```

```python
important_features_list=feats[feats['feature importance']>0].index.to_list()
important_features_list
```

```python
# adaboost with best hyperparameters
x_imp=x[important_features_list]
y=df['class']

x_train,x_test,y_train,y_test=train_test_split(x_imp,y,test_size=0.2,random_state=101)

ada_bhp=AdaBoostClassifier(n_estimators=20)
ada_bhp.fit(x_train,y_train)

train_pred=ada_bhp.predict(x_train)
test_pred=ada_bhp.predict(x_test)

print('train accuracy:',accuracy_score(y_train,train_pred))
print('test accuracy:',accuracy_score(y_test,test_pred))

score=cross_val_score(ada_bhp,x,y,cv=5)
print('cross validation score:',score.mean())
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,test_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,test_pred))
```
